```python
recency_aus = df.groupby(['customer_id'], as_index = False)['transaction_date'].max()
recency_aus.columns = ['CustomerID', 'LastPurchaseDate']
recency_aus.head()
```

```python
recent_date = recency_aus['LastPurchaseDate'].max()
```

```python
recency_aus['recency'] = recency_aus['LastPurchaseDate'].apply(lambda x: (recent_date - x).days)
recency_aus.head()
```

```python
quantiles = aus_full_rfm.quantile(q = [0.25, 0.5, 0.75])
quantiles
```

```python
quantiles.to_dict()
```

```python
def rec_score(x, m, d): # x->value, m->recency/ frequency/ monetary_value, d->quartile dict
    if x <= d[m][0.25]:
        return 4
    elif x <= d[m][0.5]:
        return 3
    elif x <= d[m][0.75]:
        return 2
    else:
        return 1

def f_m_score(x, m, d):
    if x <= d[m][0.25]:
        return 1
    elif x <= d[m][0.5]:
        return 2
    elif x <= d[m][0.75]:
        return 3
    else:
        return 4
    
    
rfm_seg = aus_full_rfm
rfm_seg['R_Quartile'] = rfm_seg['recency'].apply(rec_score, args = ('recency', quantiles,))
rfm_seg['F_Quartile'] = rfm_seg['Frequency'].apply(rec_score, args = ('Frequency', quantiles,))
rfm_seg['M_Quartile'] = rfm_seg['Monetary_value'].apply(rec_score, args = ('Monetary_value', quantiles,))
rfm_seg.head()
```

```python
rfm_seg['RFM_Score'] = rfm_seg['R_Quartile'].map(str) +\
                       rfm_seg['F_Quartile'].map(str) +\
                       rfm_seg['M_Quartile'].map(str)
rfm_seg.head()
```

```python
rfm_seg[rfm_seg.RFM_Score == '444'].sort_values('Monetary_value', ascending = False).head()
```
